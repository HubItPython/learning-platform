```python
from tkinter import *
from PIL import Image, ImageTk

def auth():
    if username.get() == "admin":
        Label(r, text="Log in Successful").grid(row=4, column=1)
        r.destroy()
        login()
    else:
        Label(r, text="Wrong username or password").grid(row=4, column=1)

def login():
    r = Tk()
    r.geometry("400x300")

    Label(r, text="First Name").grid(row=0, column=0)
    fname = Entry(r)
    fname.grid(row=0, column=1)

    Label(r, text="Last Name").grid(row=0, column=2)
    lname = Entry(r)
    lname.grid(row=0, column=3)

    Label(r, text="").grid(row=1, column=0)

    Label(r, text="Address").grid(row=2, column=2)
    address = Entry(r)
    address.grid(row=2, column=3)

    def result():
        Label(r, text=f"First Name={fname.get()}").place(x=20, y=180)
        Label(r, text=f"Last Name={lname.get()}").place(x=20, y=200)
        Label(r, text=f"Address={address.get()}").place(x=20, y=220)
        Label(r, text=f"Gender={clicked.get()}").place(x=20, y=240)

    options = ["Male", "Female"]
    Label(r, text="Gender").grid(row=2, column=0)
    clicked = StringVar()
    clicked.set("Male")
    drop = OptionMenu(r, clicked, *options).grid(row=2, column=1)

    button = Button(r, text="Submit", command=result)
    button.place(x=150, y=150)

r = Tk()
r.geometry("240x240")
r.configure(bg="#f0f2f5")

# Create a frame
frame = Frame(r)
frame.place(relx=0.1, rely=0.1, relwidth=0.8, relheight=0.8)

# Load and display an image
image = Image.open("facebook.jpg")
resize_image = image.resize((300, 106))
img = ImageTk.PhotoImage(resize_image)
label1 = Label(frame, image=img)
label1.image = img
label1.pack()

# Create username and password labels and entry widgets
Label(frame, text="Username").place(x=800, y=300)
username = Entry(frame)
username.place(x=870, y=300)

Label(frame, text="Password").place(x=800, y=340)
password = Entry(frame, show="*")
password.place(x=870, y=340)

# Create the login button
button = Button(frame, text="login", command=auth)
button.place(x=900, y=380)

# Start the main loop
r.mainloop()
```
